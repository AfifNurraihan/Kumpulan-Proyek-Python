```python
# Program Kasir Minuman Boba Sederhana
opsi = 'Ya'
while opsi == 'Ya':
    print('========================================================')
    print('xxxxxxxxxxxxxxx Selamat Datang di Boba QU xxxxxxxxxxxxxx')
    print('========================================================')
    print('                     Menu Boba AI                       ')
    print('''    1. Thai Tea Boba         Rp10000(Kecil)/Rp13000(Besar)
    2. Boba Milk Brown Sugar Rp15000(Kecil)/Rp18000(Besar)
    3. Red Velvet            Rp16000(Kecil)/Rp22000(Besar)
    4. Shimeering Galaxy     Rp18000(Kecil)/Rp25000(Besar)
    5. Kawazu Sakura         Rp15000(Kecil)/Rp18000(Besar)''')
    print('========================================================\n')
    nama_pembeli=input("Nama Pembeli        :")
    tgl=input("Tanggal Pembelian   :")
    print('========================================================')
    print('============== Silahkan Buat Pesanan Anda ==============')
    print('========================================================')
    minuman = int(input("Masukkan Kode Nomor Menu Boba yang di pesan :"))
    if minuman ==1:
        boba='Thai Tea Boba'
        ukr = str(input("Ukuran Minuman (Besar/Kecil): "))
        if ukr=='Kecil':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*10000
        elif ukr=='Besar':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*13000
    elif minuman==2:
        boba='Boba Milk Brown Sugar'
        ukr = str(input("Ukuran Minuman (Besar/Kecil): "))
        if ukr=='Kecil':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*15000
        elif ukr=='Besar':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*18000
    elif minuman==3:
        boba='Red Velvet'
        ukr = str(input("Ukuran Minuman (Besar/Kecil): "))
        if ukr=='Kecil':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*16000
        elif ukr=='Besar':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*22000
    elif minuman==4:
        boba='Shimeering Galaxy'
        ukr = str(input("Ukuran Minuman (Besar/Kecil): "))
        if ukr=='Kecil':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*18000
        elif ukr=='Besar':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*25000
    elif minuman==5:
        boba='Kawazu Sakura'
        ukr = str(input("Ukuran Minuman (Besar/Kecil): "))
        if ukr=='Kecil':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*15000
        elif ukr=='Besar':
            jml_mnm=int(input("Banyaknya pesanan : "))
            total_mnm=jml_mnm*18000
    else:
        boba='-'
        ukr='-'
        jml_mnm='-'
        total_mnm='-'
        opsi=input("Pilihan tidak tersedia, Silahkan ulangi kembali (Ya) :")
    
    def kembali(uang_pembayaran):
        if uang_pembayaran == total_mnm:
            kmbl = 0
        elif uang_pembayaran > total_mnm:
            kmbl = uang_pembayaran-total_mnm
        else:
            kmbl='-'
        return kmbl
    print('\n========================================================')
    print('================ Struk Pembelian Boba AI ===============')
    print('========================================================')
    print("Nama Pemesan      :",nama_pembeli)
    print("Tanggal Pembelian :",tgl)
    print("Menu Boba         :",boba)
    print("Jumlah Pesanan    :",jml_mnm)
    print("Ukuran Minuman    :",ukr)
    print('========================================================')
    print("Total Pembayaran  : Rp",total_mnm)
    uang_pembayaran = int(input("Uang Pembayaran   : Rp "))
    print('========================================================')
    print('Uang Kembali      : Rp',kembali(uang_pembayaran))
    print('========================================================')
    print('******************* Selamat Menikmati ******************')
    print('========================================================\n\n')
    opsi=input("Mau memesan lagi? Ketik Ya/Tidak : ")
```
